```python
emoji = re.compile("["
                   u"\U0001F600-\U0001F64F"  #Emoticons
                   u"\U0001F300-\U0001F5FF"  #Symbols & pictographs
                   u"\U0001F680-\U0001F6FF"  #Transport & map symbols
                   u"\U0001F1E0-\U0001F1FF"  #Flags (iOS)
                   u"\U00002500-\U00002BEF"  #Chinese char
                   u"\U00002702-\U000027B0"
                   u"\U00002702-\U000027B0"
                   u"\U000024C2-\U0001F251"
                   u"\U0001f926-\U0001f937"
                   u"\U00010000-\U0010ffff"
                   u"\u2640-\u2642"
                   u"\u2600-\u2B55"
                   u"\u200d"
                   u"\u23cf"
                   u"\u23e9"
                   u"\u231a"
                   u"\ufe0f"  #Dingbats
                   u"\u3030"
                   "]+", flags = re.UNICODE) #Flag option as Unicode


def cleanText1(text):
    newText = re.sub(r'(http|ftp|https)\S+', '', text) # remove url
    # there's still have https://t.co/DPAaiX18tD
    return newText

def cleanText2(text):
    newText = re.sub(emoji, '', str(text)) # remove emoji
    # newText = re.sub(r'#([a-zA-Z0-9ก-๙_]+)', '', newText) # remove hashtag
    # newText = newText.translate(str.maketrans('', '', string.punctuation)) # remove punctuation
    newText = pythainlp.util.normalize(newText) # remove repeated letter
    newText = ' '.join(newText.split()) # keep only one white space
    return newText
```

```python
def drop_and_display_tweets_by_word_from_list(df, wordlist):
    num_deleted_tweet = 0
    for word in wordlist:
        print("{} : {} tweets".format(word, df['text'].str.contains(word, case=False).sum()))
        num_deleted_tweet += df['text'].str.contains(word, case=False).sum()
        print(df[df['text'].str.contains(word, case=False)]['text'])
        df = df[~df['text'].str.contains(word, case=False)]
    print("total deleted tweets from this list : {}".format(num_deleted_tweet))
    print('-----------------------------------------')
    return df
```

```python
def drop_tweets_by_word_from_list(df, wordlist):
    num_deleted_tweet = 0
    for word in wordlist:
        condition = df['clenedText'].str.contains(word, case=False)
        num_tweets_contain = condition.sum()
        num_deleted_tweet += num_tweets_contain
        df = df[~df['clenedText'].str.contains(word, case=False)]
        print("{} : {} tweets".format(word, num_tweets_contain))
    print('-----------------------------------------')
    print("total deleted tweets from this list : {}".format(num_deleted_tweet))
    print('-----------------------------------------')
    return df
```

```python
def findall_hashtag(tokens):
    hashtags = []
    for token in tokens:
        if token.startswith("#"):
            hashtags.append(token)
    return hashtags
```

```python
def remove_emoji(text):
  return ''.join([c for c in text if c not in emoji.UNICODE_EMOJI])
```

```python
def compute_ngram(text_tokens, n):
    tokens = []
    if len(text_tokens) >= n:
        for i in range(len(text_tokens)):
            token_list = []
            if i == len(text_tokens) - (n-1):
                break
            for j in range(n):
                if i+j >= len(text_tokens):
                    break
                if text_tokens[i+j] == ' ':
                    continue
                else: token_list.append(text_tokens[i+j])
            ngram = ' '.join(token_list) + ', ' + ''.join(token_list)
            tokens.append(ngram)
        return tokens
    else:
        print(text_tokens)
        return "this tweet's shorter than n_gram"
```

```python
def keep_tweets_by_wordlist(df_origi, df_append, query_list):
    num_keep_tweets = 0
    for word in query_list:
        print("{} : {} tweets".format(word, str(df_origi['clenedText'].str.contains(word, case=False).sum())))
        num_keep_tweets += df_origi['clenedText'].str.contains(word, case=False).sum()
        df_append = df_append.append(df_origi[df_origi['clenedText'].str.contains(word, case=False)], ignore_index=True)
        df_origi.drop(df_origi[df_origi['clenedText'].str.contains(word, case=False)].index, inplace=True)
    print('-----------------------------------------')
    print("total appened tweets from this query list : {}".format(num_keep_tweets))
    print("total appened tweets : {}".format(df_append.shape[0]))
    print("remain tweets : {}".format(df_origi.shape[0]))
    print('-----------------------------------------')
    return df_origi, df_append
```

```python
def keep_tweets_by_tokens(df_origi, df_append, tokens_query_list):
    num_keep_tweets = 0
    for token_list in tokens_query_list:
        condition = df_origi.apply(lambda row: all(token in row['tokens'] for token in token_list), axis=1)
        num_ = df_origi.apply(lambda row: all(token in row['tokens'] for token in token_list), axis=1).sum()
        print("{} : {} tweets".format(token_list, num_))
        num_keep_tweets += num_
        df_append = df_append.append(df_origi[condition], ignore_index=True)
        df_origi.drop(df_origi[condition].index, inplace=True)
    print('-----------------------------------------')
    print("total appened tweets from this query list : {}".format(num_keep_tweets))
    print("total appened tweets : {}".format(df_append.shape[0]))
    print("remain tweets : {}".format(df_origi.shape[0]))
    print('-----------------------------------------')
    return df_origi, df_append
```

```python
def keep_tweets_by_tokens_inorder(df_origi, df_append, tokens_query_list):
    num_keep_tweets = 0
    for token_list in tokens_query_list:
        condition = df_origi['tokens'].apply(lambda x: any([all(x.index(i) < x.index(j) for i, j in zip(token_list, token_list[1:]))] \
                                                 if all(i in x for i in token_list) else [False]))
        num_ = df_origi.apply(lambda row: all(token in row['tokens'] for token in token_list), axis=1).sum()
        print("{} : {} tweets".format(token_list, num_))
        num_keep_tweets += num_
        df_append = df_append.append(df_origi[condition], ignore_index=True)
        df_origi.drop(df_origi[condition].index, inplace=True)
    print('-----------------------------------------')
    print("total appened tweets from this query list : {}".format(num_keep_tweets))
    print("total appened tweets : {}".format(df_append.shape[0]))
    print("remain tweets : {}".format(df_origi.shape[0]))
    print('-----------------------------------------')
    return df_origi, df_append
```

```python
def most_frequent_token_combinations(df, n):
    df['combinations'] = df['tokens'].apply(lambda x: list(set(itertools.combinations(x, n))))
    combination_count = Counter([item for sublist in df['combinations'] for item in sublist])
    return combination_count
```
